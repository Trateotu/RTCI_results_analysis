```python
import os
import pyparsing as pp
import pandas as pd
```

```python
main_path = 'Amsterdam_output'
output_path = os.path.join(main_path,"parsed")
if not os.path.isdir(output_path):
    os.mkdir(output_path)
```

```python
dfs = dict()
failed = list()
for dat in os.listdir(main_path):
    if dat.endswith(".dat"):
        try:
            df = pd.read_table(os.path.join(main_path,dat))
            dfs[dat[:-4]]= df
            df.to_csv(os.path.join(output_path,dat).replace(".dat",".csv"))
            print("Loaded {} records on {} columns from {} file".format(df.shape[0],df.shape[1],dat))
        except:
            failed.append(dat)
            print("Failed to load: ",dat)
```

```python
nested_braces = pp.nestedExpr('{', '}')
f = open(os.path.join(main_path,'o_selected_paths.dat'))
t = list()
flag = True
i=0
for l in f.readlines():
    i+=1
    if divmod(i,1000)[1] == 0:
        break
    
    if flag:
        headers = l.split("\t")
        flag = False
    else:
        t.append(nested_braces.parseString("{"+l+"}").asList()[0])
h = headers[:-1]
h.append('Unknown1')
h.append('Unknown2')
o_selected_paths = pd.DataFrame.from_records(t, columns = h)
o_selected_paths
```
